## Retail and food services sales, total Query
Note the reformatting of the date so as to convert it to an actual date data type so it could be ordered properly to show the trend, which is increasing over time. The years that I used are from 2001-2021. It is interesting to note the spike in activity in December and then a steep decline in the trend in January and February before it reverses course and heads in a positive direction once more. However, an exception to this occured starting in March 2020 for approximately 3-months where the totals for these months took a dip due to the slowdown in dining out that occurred during these months. 

Considering the data for the total sales for the retail and food services categories, what is the trend of these retail categories? Can this data be displayed clearly, or do you need to adjust some parameters to reduce extraneous details and be able to visualize a clean trend? --> Once the dates were converted to a SQL date data type, the __'Retail and food services sales, total'__ trend is clearly increasing over time with the given years (2001 - 2021). 

```python
import mysql.connector
import matplotlib.pyplot as plt

conn = mysql.connector.connect(user='root', 
       password='password',
       host='127.0.0.1',
       database='mrts_db')
cursor = conn.cursor()

sql = """
SELECT 
    CASE
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Jan' 
            THEN STR_TO_DATE(CONCAT("January", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Feb' 
            THEN STR_TO_DATE(CONCAT("February", "28", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Mar' 
            THEN STR_TO_DATE(CONCAT("March", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")            
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Apr' 
            THEN STR_TO_DATE(CONCAT("April", "30", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")        
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'May' 
            THEN STR_TO_DATE(CONCAT("May", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")            
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Jun' 
            THEN STR_TO_DATE(CONCAT("June", "30", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Jul' 
            THEN STR_TO_DATE(CONCAT("July", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")            
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Aug' 
            THEN STR_TO_DATE(CONCAT("August", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Sep' 
            THEN STR_TO_DATE(CONCAT("September", "30", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Oct' 
            THEN STR_TO_DATE(CONCAT("October", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Nov' 
            THEN STR_TO_DATE(CONCAT("November", "30", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        WHEN SUBSTRING_INDEX(REPLACE(`Month and Year`, '.', ''),' ',1) = 'Dec' 
            THEN STR_TO_DATE(CONCAT("December", "31", SUBSTRING(REPLACE(`Month and Year`, '.', ''), 5)), "%M %D %Y")
        ELSE REPLACE(`Month and Year`, '.', '')
        END AS month_year,
CAST(REPLACE(`Retail and food services sales, total`, ',', '') AS UNSIGNED) 
FROM mrts_table
ORDER BY month_year;
"""

cursor.execute(sql)

month = []
sales = []

for row in cursor.fetchall():
    print(row)
    month.append(row[0])   
    sales.append(row[1])
    
cursor.close()
conn.close()

plt.figure(figsize=(18,14))
plt.plot(month, sales)
plt.xlabel('Month and Year')
plt.ylabel('Monthly Sales')
plt.show()

                              
```

# Trend Analysis For: bookstores, sporting goods stores, and hobbies, toys, and games stores

When comparing businesses such as bookstores, sporting goods stores, and hobbies, toys, and games stores, which is the highest trend of all of these options? Which one grew faster? Which one is higher? Is there a seasonal pattern? Are there any changes in 2020? Which is better, monthly or yearly?

For the bookstores, there is a clear seasonal pattern with December and January logging in the highest sales, as well as August, which is typically when the school season starts. However, the overall sales trend for bookstores is in a decline for the years of 2001 to 2021. Again, notice the steep dip in months of March, April, and May in 2020.

For Sporting Goods Stores, the highest sales occur in December and then have a steep decline in January. The overall trend appears to be a slight increase. This appears to have the highest trend among the three types of businesses.

For the Hobby, Toy, Game Stores business type, sales once again are strongest in December, coinciding with Christmas. However the trend appears to be the weakest among the three business segments with perhaps a very slight increase in trend but it is not too apparent.
